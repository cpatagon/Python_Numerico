## 2. Iteración `while`

```
b = 1
while b <= 5:
    print (b)
    b *= 2
```

```
b = 1
while b < 5:
    print (b)
    b += 1
```

## 4. Controles `break` y `continue`

```
for letra in 'esmerilemelo':     
    if letra == 'i':
        break
    print ('letra actual :', letra)
```

```
for letra in 'esmerilemelo':     
    if letra == 'i':
        continue
    else:
        print ('letra actual :', letra)
```

## 5. Controles `try` y `except`

```
for n in range(6)[::-1]:
    print (n)
```

```
for n in range(6)[::-1]:
    val = 5/n
    print ('5/{}={}'.format(n,val))
```

```
for n in range(6)[::-1]:
    val = 5/float(n)
    print ('5/{}={}'.format(n,val))
```

```
for n in range(6)[::-1]:
    try:
        val = 5/float(n)
        print ('5/{}={}'.format(n,val))
    except:
        print ('5/0 es infinito')
```

...la mejor practica es incluir el nombre de la excepción

```
for n in range(6)[::-1]:
    try:
        val = 5/float(n)
        print ('5/{}={}'.format(n,val))
    except ZeroDivisionError:
        print ('5/0 es infinito')
```

```
for n in range(6)[::-1]:
    try:
        val = 5/float(n)
        print ('5/{}={:2.1f}'.format(n,val))
    except ZeroDivisionError:
        print ('5/0 es infinito')
```
