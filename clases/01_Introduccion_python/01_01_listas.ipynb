# 1 listas en Python

Las listas corresponde a una secuencia de datos que se escriben entre **corchetes y se separan por comas**.  Nos permiten almacenar cualquier tipo de valor como enteros, cadenas y hasta otras funciones; por ejemplo: Las listas pueden ser numéricas:



Usando el contructor  **list()** se puede hacer una lista:

```
a=("apple", "banana", "cherry")  // esta es una tupla
print(a)
print(type(a))
thislist = list(a) # note the double round-brackets
print(thislist)
print(type(thislist))
```

```
foo1 = [1,2,3,4]
print (foo1)
```

```
yf = [cx*2 for cx in foo1]
print(yf)
```

Notar que con el comando type podemos identificar el tipo de estructura de datos.



```
type(foo1)
```

Las listas pueden tener string o texto el que se escribe entre medio de comillas

```
foo2 = ['a','b','c','d']
print (foo2)
```

O pueden ser una mezcla de distintos tipo o type()

```
foo3 = ['a',33, 1.4]
print (foo3)
print(type(foo3))
for a in foo3:
  print('la variable \'{0}\' es del tipo {1}'.format(a, type(a)))
```

Para declarar formalmente una lista ocupamos la función list() y luego los valores que crearemos:



```
foo4 = list([1,2])
print (foo4)
```

Tambien podemos crear una lista vacia



```
foo_empty2 = []
foo_empty3 = list()
print (foo_empty2, foo_empty3)
```

### Creación de listas de datos con range()

Con la función range podemos crear una secuencia de números en donde el valor que está al interior de la funcióón nos especifica su tamaño.

```
nums = list(range(20))
print(nums)
```

```
for a in range(5):
  print(a)
```

```
list(range(5))
```

Notar que los valores parten de 0 y terminan es 19. Es decir son 20 valores que comienzan desde 0. Si especificamos 2 valores a la función, separados por coma, el prmer valor nos indicará el comienzo y el segundo valor el anterior al final.

```
nums = list(range(3,7))
print (nums)
```

```
nums = list(range(1,5))
print (nums)
```

```
 Una tercera modalidad será con tres valores en donde el último valor corresponde al paso que se ejecutara entre cada valor
```

```
nums = list(range(0,100,5))
print (nums)
```

Por último podemos crear listas dentro de listas ocupando tambien el la función  range()

```
m=[list(range(10)), list(range(2))]
print (m)
```

## 1.3. Métodos

### Lista de Métodos

Python tiene un conjunto de métodos incorporados que puedes usar en listas.

Método | Descripción
--- | ---
**append()** | Añade un elemento al final de la lista
**clear()** | Elimina todos los elementos de la lista
**copy()** | Devuelve una copia de la lista
**count()** | Devuelve el número de elementos con el valor especificado
**extend()** | Añade los elementos de una lista (o cualquier iterable) al final de la lista actual
**index()** | Devuelve el índice del primer elemento con el valor especificado
**insert()** | Añade un elemento en la posición especificada
**pop()** | Elimina el elemento en la posición especificada
**remove()** | Elimina el elemento con el valor especificado
**reverse()** | Invierte el orden de la lista
**sort()** | Ordena la lista
