## 1. Crear diccionarios

Los diccionarios en Python son un tipo de estructuras de datos que permite guardar un conjunto no ordenado de pares clave-valor, siendo las claves únicas dentro de un mismo diccionario (es decir que no pueden existir dos elementos con una misma clave)

```
foo = {'lat':38, 'lon':32}
print (foo)
```

```
print (foo['lon'])
```

```
print (foo.get('lon'))
```

```
print (foo['lat'])
```

```
foo = {'lat':38, 'lon':32, 0:'qwerty'}
print (foo)
```

```
print (foo[0])
```

```
foo = dict(lat=38,lon=32)
print (foo)
```

```
foo = dict(lat=38,lon=32,oso='querty')
print (foo)
```

cambiando el valor de un valor 

```
foo['lat']=75
print(foo)
```

## 2. Diccionarios embebidos

```
foo = {'lat':38, 'lon':32, 0:'qwerty',
      1:{'nombre':'santiago','altitud':623.8}}
print(foo)
```

```
foo = {'lat':38, 'lon':32, 0:'qwerty',
      1:{'nombre':'santiago','altitud':623.8},'rango':list(range(10))}
print (foo)
```

## 3. Métodos

```
print (foo.keys())
```

```
print (foo.values())
```

```
for k in foo.keys():
    print (k)
```

```
print (list(foo.values()))
```

```
print (len(foo))
```

```
for key in foo.keys():
  value=foo[key]
  print ('valor de {} es {}'.format(key, value))
```
