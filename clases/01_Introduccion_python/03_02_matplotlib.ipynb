# Introducción a Matplotlib

Matplotlib es una biblioteca en Python utilizada para crear gráficos y visualizaciones. 

Es muy poderosa y flexible, lo que la hace ideal para generar gráficos de alta calidad de manera simple y eficiente. 

En esta clase, se describirá las bases de Matplotlib y aprenderemos a crear figuras, personalizar gráficos y trabajar con subplots.

## Matplotlib

```python
import matplotlib.pyplot as plt
```

## 1. Figuras y Axes
Las figuras en Matplotlib actúan como contenedores donde se dibujan uno o más gráficos. Dentro de una figura, los 'axes' son los espacios donde realmente se trazan los datos. A continuación, veremos cómo crear una figura y agregarle axes.

fig = plt.figure()  # Creamos una figura
plt.show()  # Mostramos la figura vacía

```python
ax = plt.axes()  # Creamos un objeto axes
plt.show()  # Mostramos los axes
```

### Ejemplo: Personalización de una Figura
Podemos personalizar la figura estableciendo el tamaño y el color de fondo, entre otras cosas.

```python
fig = plt.figure(figsize=(8, 6), facecolor='lightblue')  # Tamaño y color de fondo de la figura
ax = plt.axes()
plt.show()
```

## 2. Subplots
Los subplots permiten crear múltiples gráficos dentro de la misma figura. Esto es útil cuando queremos comparar diferentes conjuntos de datos o mostrar información relacionada en un mismo lugar.

```python
fig, axs = plt.subplots(2, 2)  # Creamos una figura con una cuadrícula de 2x2 subplots
plt.show()
```

### Ejemplo: Personalización de Subplots
Podemos ajustar los subplots individualmente para diferentes tipos de gráficos.

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot([1, 2, 3], [4, 5, 6],'r')  # Primer subplot
axs[0, 1].plot([1, 2, 3], [6, 5, 4],'b')  # Segundo subplot
axs[1, 0].plot([1, 2, 3], [4, 4, 4],'y')  # Tercer subplot
axs[1, 1].plot([4, 9, 7], [6, 4, 2],'g')  # Cuarto subplot
plt.show()
```

## 3. subplots

```python
x = np.arange(0.0, 5.0, 0.1)
pi = np.pi
func1 = lambda x: np.exp(-x) * np.cos(2*pi*x)
func2 = lambda x: np.cos(2*pi*x)
y1 = func1(x)
y2 = func2(x)

# creamos figuras
fig, axs = plt.subplots(2,1)
axs[0].plot(x,y1,'-o')
axs[1].plot(x,y2,'-o',color='r')
plt.show()
```

```python
fig, axs = plt.subplots(2,1, sharex=True)  # comparte x-axis
axs[0].plot(x,y1,'-o')
axs[1].plot(x,y2,'-o',color='r')
plt.subplots_adjust(hspace=0)
plt.show()
```

```python
# comparte x-axis e y-axis
fig, axs = plt.subplots(2,2, sharex=True, sharey=True)  

axs[0,0].plot(x,y1,':', color='g')
axs[0,1].plot(x,y1,'-.', color='b')
axs[1,0].plot(x,y1,'-d', color='y')
axs[1,1].plot(x,y1,'--', color='r')

axs[0,0].set_ylabel('ylabel')
axs[1,0].set_xlabel('xlabel')

plt.show()
```

```python
# aumenta figsize
fig, axs = plt.subplots(2,2, figsize=(10,8), 
                        sharex=True, sharey=True)  

axs[0,0].plot(x,y1,':', color='g')
axs[0,1].plot(x,y1,'-.', color='b')
axs[1,0].plot(x,y1,'-d', color='y')
axs[1,1].plot(x,y1,'--', color='r')

axs[0,0].set_ylabel('ylabel', fontsize=15)
axs[1,0].set_xlabel('xlabel', fontsize=15)

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,8), 
                        sharex=True, sharey=True)

axs = axs.flatten() # <- utilizamos 1D array en lugar de 2D
axs[0].plot(x,y1,':', color='g')
axs[1].plot(x,y1,'-.', color='b')
axs[2].plot(x,y1,'-d', color='y')
axs[3].plot(x,y1,'--', color='r')

axs[0].set_ylabel('ylabel', fontsize=15)
axs[2].set_xlabel('xlabel', fontsize=15)

plt.tight_layout()
plt.show()
```

### 4. 2D plot

```python
func = lambda x,y: (1 - x/2 + x**5 + y**3)*np.exp(-x**2 - y**2)
```

```python
# creamos grilla
n = 300
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x,y)
Z = func(X, Y)
```

** `Contourf` y `contour` **

```python
fig, ax = plt.subplots()
ax.contourf(X, Y, Z, 8, alpha=.75, cmap='RdBu')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.contourf(X, Y, Z, 8, alpha=.75, cmap='RdBu')
ax.contour(X, Y, Z, 8, colors='black', linewidth=.5)
plt.show()
```

```python
fig, ax = plt.subplots()

cf = ax.contourf(X, Y, Z, 8, alpha=.75, cmap='RdBu')
c = ax.contour(X, Y, Z, 8, colors='black', linewidth=.5)

ax.set_xlabel('X')
ax.set_ylabel('Y')

# agrega etiquetas de contorno
plt.clabel(c, fontsize=10)

# agrega colorbar
plt.colorbar(cf, label='[unidades]') # <- debemos 
                                     # entregarle un 'mappable'

plt.show()
```

**`imshow`**

```python
# destinado para imágenes, no considera grilla
fig, ax = plt.subplots()
im = ax.imshow(Z)
plt.colorbar(im)
plt.show()
```

**`pcolor` y `pcolormesh`**

```python
# considera grilla
fig, ax = plt.subplots()
pc = ax.pcolor(X, Y, Z)
plt.colorbar(pc)
plt.show()
```

```python
fig, ax = plt.subplots()
pc = ax.pcolormesh(X,Y,Z)
plt.colorbar(pc)
plt.show()
```

diferencia pcolor y pcolormesh:
pcolormesh es más rápido

diferencia imshow y pcolormesh:
*imshow sigue convención de procesamiento de imágenes: origen en esquina superior izquierda; pcolormesh usa esquina inferior izquierda o definido por grilla.

*posicion de pixel: imshow centrado y pcolormesh entre valores de grilla


**`quiver`**

```python
fig, ax = plt.subplots()
n = 8
X,Y = np.mgrid[0:n,0:n]
T = np.arctan2(Y-n/2.0, X-n/2.0)
R = 10+np.sqrt((Y-n/2.0)**2+(X-n/2.0)**2)
U,V = R*np.cos(T), R*np.sin(T)
ax.quiver(X,Y,U,V)
plt.show()
```

```python
fig, ax = plt.subplots()
n = 8
X,Y = np.mgrid[0:n,0:n]
T = np.arctan2(Y-n/2.0, X-n/2.0)
R = 10+np.sqrt((Y-n/2.0)**2+(X-n/2.0)**2)
U,V = R*np.cos(T), R*np.sin(T)
qv = ax.quiver(X,Y,U,V,R)
plt.colorbar(qv)
plt.show()
```
