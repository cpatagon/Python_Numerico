# Derivada numérica
## Motivación

La diferenciación es una  herramienta esencial en nuestra profesión y del cálculo numérico. 

diferenciar significa "marcar diferencia; distinguir ..." 

En matemáticas se representa como la razón de cambio de una variable dependiente respecto a una variable independiente.

$
\frac{\Delta x}{\Delta y} =  \frac{f(x_i + \Delta x)- f(x_i)}{\Delta x}
$

si $\Delta x \rightarrow 0$ entonces las diferencias se convierten en una derivada

##  Derivada formal
 Entonces nuestra razón convierte en una derivada

$
\frac{dx}{dy} =  \lim_{\Delta x \to 0} \frac{f(x_i + \Delta x)- f(x_i)}{\Delta x}
$

Y gráficamente la derivada se puede ver como la pendiente de la recta tangente a la curva en $x_i$ 


## Otras derivadas

**segunda derivada:**  es la derivada de la primera derivada y expresa la rapidez del cambio.

**derivadas parciales:** funciones que dependen de mas de una variable 

## formas comunes de derivadas

**función simple de deribar:** como polinomial, exponencial o trigonométrica. Estas se pueden evaluar analíticamente sin necesidad de de un método numérico.

**función compleja:**   o imposible de derivar directamente. En este caso podemos ocupar métodos numéricos para su solución aproximada

**tabulada:** en que la función se presenta en datos discretos de la forma $x, f(x)$, por ejemplo cuando se tienen datos experimentales o de campo.  Sólo puede ser resulta mediante métodos numéricos.


## Diferenciación por áreas iguales
	
  Se emplea una diferencia dividida simple ($\Delta y /\Delta x$) para estimar la pendiente. 
	
* a partir de los datos se calcula $\Delta y /\Delta x$

* se dibuja la solución la que quedara como una curva escalonada

* se busca el centro de los escalones y se traza por ellos una curva que los atraviese, suavizando con ello la solución.

## Funciones en Python
	
**numpy.diff()[source]** Derivada aproximada. devuelve un vector de  n-1 datos 
	**np.gradient(f)** El gradiente se calcula utilizando diferencias centrales precisas de segundo orden en los puntos interiores y diferencias precisas de un lado (hacia adelante o hacia atrás) de primer o segundo orden en los límites. Por tanto, el gradiente devuelto tiene la misma forma que la matriz de entrada.
 

```
import numpy as np

f = np.array([1, 2, 4, 7, 11, 16])
print(np.gradient(f))

print(np.gradient(f, 2))
```
