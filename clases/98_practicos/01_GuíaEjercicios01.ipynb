```python
import random
import numpy as np
import matplotlib.pyplot as plt

def aproximar_pi(n):
    dentro_circulo = 0
    total_puntos = n
    
    for _ in range(total_puntos):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if x*x + y*y <= 1:
            dentro_circulo += 1
    
    return 4 * dentro_circulo / total_puntos

# Prueba la función
n = 1000000
print(f"Aproximación de π con {n} puntos: {aproximar_pi(n)}")
```
