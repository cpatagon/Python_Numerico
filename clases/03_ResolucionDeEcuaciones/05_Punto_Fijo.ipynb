# Método de Iteración Simple de Punto Fijo

## Descripción General

El método de iteración simple de punto fijo es una técnica numérica utilizada para encontrar las raíces de una función. Es un método abierto, lo que significa que no requiere un intervalo inicial que contenga la raíz, sino que parte de un único valor inicial.

## Fundamento Teórico

1. Se parte de una ecuación $f(x) = 0$.
2. Se reordena la ecuación en la forma $x = g(x)$.
3. Se utiliza la fórmula iterativa: $x[i+1] = g(x[i])$

donde $x[i]$ es el valor actual y $x[i+1]$ es la siguiente aproximación.

## Escoger $g(x)$

- **Paso 1: $f(x) = 0$**

Este paso se refiere a la ecuación original que queremos resolver. Aquí, f(x) es una función matemática y estamos buscando los valores de x que hacen que esta función sea igual a cero. Estos valores son las raíces o soluciones de la ecuación.

Ejemplo: Supongamos que queremos resolver la ecuación $x² - 4 = 0$. En este caso, $f(x) = x² - 4$.

- **Paso 2: Reescribir como $x = g(x)$**

Este paso es crucial y a menudo requiere algo de creatividad. Debemos transformar la ecuación original en una forma donde x esté sola en un lado de la igualdad, y el otro lado sea una función de x. Esta nueva función g(x) se llama "función de iteración".
Siguiendo con nuestro ejemplo $x² - 4 = 0$, podríamos reescribirla de varias formas:

 $$ x = \sqrt{x^2 - 4 + 4} = \sqrt{x^2 + 0} = \sqrt{x^2} = |x|$$
 Aquí, $ g(x) = |x|$
$$ x = \sqrt{4} = 2 $$
Aquí, $g(x) = 2$
$$ x = \frac{x^2 + 4}{x} $$
Aquí, $g(x) = \frac{x^2 + 4}{x}$

La elección de $g(x)$ es importante porque afecta la convergencia del método. Algunas formas de g(x) convergerán rápidamente, otras lentamente, y algunas pueden no converger en absoluto




## Características Principales

- Es un método abierto que puede converger más rápidamente que los métodos cerrados.
- Puede diverger en algunos casos, alejándose de la raíz verdadera.
- Cuando converge, generalmente lo hace de manera lineal.

## Criterio de Convergencia

El método converge si |g'(x)| < 1 en la vecindad de la raíz, donde g'(x) es la derivada de g(x).

## Representación Gráfica

Se puede visualizar gráficamente mediante el método de las dos curvas:
1. Se grafica y1 = x
2. Se grafica y2 = g(x)
3. La intersección de estas curvas representa la raíz.

## Pseudocódigo Detallado

```
Función IteracionPuntoFijo(g, x0, tolerancia, max_iteraciones)
    // g: función g(x) en la forma x = g(x)
    // x0: valor inicial
    // tolerancia: error máximo aceptable
    // max_iteraciones: número máximo de iteraciones permitidas

    x_actual = x0
    iteracion = 0

    Mientras iteracion < max_iteraciones:
        x_siguiente = g(x_actual)
        error = |x_siguiente - x_actual| / |x_siguiente|

        Si error < tolerancia:
            Retornar x_siguiente  // Convergencia alcanzada
        
        x_actual = x_siguiente
        iteracion = iteracion + 1

        // Opcional: imprimir estado actual
        Imprimir("Iteración:", iteracion, "x =", x_actual, "Error =", error)

    // Si se llega aquí, no se alcanzó la convergencia
    Lanzar Error("El método no convergió en el número máximo de iteraciones")

Fin Función
```

### Explicación del Pseudocódigo:

1. La función toma como entrada:
   - g(x): la función reordenada en la forma x = g(x)
   - x0: un valor inicial para comenzar la iteración
   - tolerancia: el error máximo aceptable
   - max_iteraciones: límite de iteraciones para evitar bucles infinitos

2. Inicializamos x_actual con el valor inicial x0.

3. Entramos en un bucle que continúa hasta alcanzar la convergencia o el número máximo de iteraciones:
   - Calculamos el siguiente valor x_siguiente usando g(x_actual)
   - Calculamos el error relativo
   - Si el error es menor que la tolerancia, hemos convergido y retornamos el resultado
   - Si no, actualizamos x_actual y continuamos con la siguiente iteración

4. Si el bucle termina sin convergencia, lanzamos un error.

5. Opcionalmente, podemos imprimir el estado en cada iteración para seguir el progreso del método.

## Ventajas y Desventajas

Ventajas:
- Simple de implementar y entender.
- Puede ser muy eficiente si la función g(x) está bien elegida.

Desventajas:
- Puede diverger si no se cumple el criterio de convergencia.
- La velocidad de convergencia puede ser lenta en comparación con otros métodos más avanzados.

## Aplicaciones

- Resolución de ecuaciones no lineales.
- Problemas de ingeniería donde se busca un punto de equilibrio.
- Como paso intermedio en métodos más complejos.

## Ejemplo de Implementación en Python



````python
import numpy as np
def punto_fijo(g, x0, tol=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        x_new = g(x)
        if abs((x_new - x) / x_new) < tol:
            return x_new
        x = x_new
        print(f"Iteración {i+1}: x = {x}, Error = {abs((x_new - x) / x_new)}")
    raise ValueError("El método no convergió")

# Ejemplo de uso
import math

def g(x):
    return math.exp(-x)  # Para resolver e^(-x) = x

try:
    raiz = punto_fijo(g, 0.5)
    print(f"La raíz aproximada es: {raiz}")
except ValueError as e:
    print(e)

````

Este método es fundamental en el estudio de los métodos numéricos y sirve como base para entender métodos más avanzados de búsqueda de raíces.

````python
import numpy as np
import matplotlib.pyplot as plt

def g(x):
    return np.exp(-x)

def plot_fixed_point_iteration(g, x0, num_iterations):
    x = np.linspace(0.4, 0.7, 100)
    
    plt.figure(figsize=(10, 8))
    plt.plot(x, g(x), 'r-', label='y = g(x)')
    plt.plot(x, x, 'b-', label='y = x')
    
    x_n = x0
    for i in range(num_iterations):
        y_n = g(x_n)
        plt.plot([x_n, x_n], [x_n, y_n], 'g--')
        plt.plot([x_n, y_n], [y_n, y_n], 'g--')
        plt.plot(x_n, y_n, 'go')
        x_n = y_n
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Método de Iteración Simple de Punto Fijo')
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()

# Ejemplo de uso
plot_fixed_point_iteration(g, 0.5, 5)
````
