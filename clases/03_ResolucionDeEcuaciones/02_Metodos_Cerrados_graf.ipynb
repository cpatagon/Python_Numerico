# Métodos Cerrados para Encontrar Raíces de Ecuaciones

## Introducción

Los métodos cerrados, también conocidos como **métodos de intervalos**, son técnicas fundamentales en el cálculo numérico para encontrar raíces de ecuaciones. 

Estos métodos se basan en el principio de que si **una función continua cambia de signo en un intervalo, debe existir al menos una raíz** dentro de ese intervalo.



### Contexto histórico

La búsqueda de raíces de ecuaciones es un problema antiguo en matemáticas. Los métodos cerrados, como el método de bisección, tienen sus orígenes en técnicas de aproximación utilizadas por matemáticos desde la antigüedad. 

Sin embargo, su formalización y análisis riguroso se desarrollaron principalmente en los siglos XVIII y XIX, con el avance del análisis matemático.

### Aplicaciones prácticas

En ingeniería y ciencias aplicadas, los métodos cerrados son esenciales para resolver problemas que involucran:

1. Diseño de estructuras y componentes mecánicos
2. Análisis de circuitos eléctricos no lineales
3. Problemas de equilibrio químico
4. Optimización de procesos industriales
5. Modelado de sistemas dinámicos en física e ingeniería

## Fundamentos teóricos

### Teorema del valor intermedio

El fundamento teórico de los métodos cerrados es el teorema del valor intermedio, que establece:

Si $f(x)$ es una función continua en el intervalo cerrado $[a,b]$, y $f(a)$ y $f(b)$ tienen signos opuestos, entonces existe al menos un valor $c$ en $(a,b)$ tal que $f(c) = 0$.

Matemáticamente:

$$ \text{Si } f(a) \cdot f(b) < 0, \text{ entonces } \exists c \in (a,b) : f(c) = 0 $$

### Método gráfico

El método gráfico es una técnica visual para aproximar las raíces de una ecuación. Consiste en:

1. Graficar la función $f(x)$
2. Identificar los puntos donde la gráfica cruza el eje x
3. Estimar las coordenadas x de estos puntos de cruce


![](./img/cambio_signo.png)

### Principios de los métodos cerrados

1. Requieren un intervalo inicial $[a,b]$ donde $f(a)$ y $f(b)$ tengan signos opuestos.
2. En cada iteración, reducen sistemáticamente el tamaño del intervalo que contiene la raíz.
3. Garantizan encontrar una raíz si existe en el intervalo inicial (convergencia garantizada).

## Implementación en Python

Veamos una implementación básica del método gráfico en Python utilizando matplotlib:


```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function(f, a, b, n=1000):
    """
    Grafica una función f(x) en el intervalo [a,b].
    
    Parámetros:
    f : función a graficar
    a, b : límites del intervalo
    n : número de puntos para la gráfica
    """
    x = np.linspace(a, b, n)
    y = f(x)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='f(x)')
    plt.axhline(y=0, color='r', linestyle='--', label='y=0')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Gráfica de f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Ejemplo de uso
def f(x):
    return np.sin(10*x) + np.cos(3*x)

plot_function(f, 0, 5)
```

Este código genera una gráfica de la función $f(x) = \sin(10x) + \cos(3x)$ en el intervalo $[0, 5]$, permitiendo una visualización de las posibles raíces.

## Ejemplo práctico

Problema: Un ingeniero está diseñando un sistema de amortiguación y necesita encontrar el punto de equilibrio de un resorte no lineal. La fuerza del resorte está dada por la ecuación:

$$ f(x) = 2x^3 - 12x + 10 $$

donde $x$ es el desplazamiento en metros y $f(x)$ es la fuerza en Newtons. Encuentra gráficamente los puntos de equilibrio (raíces) en el intervalo $[-3, 3]$.

Solución:

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2*x**3 - 12*x + 10

def plot_spring_force(a, b, n=1000):
    x = np.linspace(a, b, n)
    y = f(x)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='f(x) = 2x³ - 12x + 10')
    plt.axhline(y=0, color='r', linestyle='--', label='Equilibrio')
    plt.xlabel('Desplazamiento (m)')
    plt.ylabel('Fuerza (N)')
    plt.title('Fuerza del resorte no lineal')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_spring_force(-3, 3)
```

Esta gráfica muestra tres puntos donde la función cruza el eje x, indicando tres puntos de equilibrio. Las raíces aproximadas son:

1. x ≈ -2.5
2. x ≈ 0.5
3. x ≈ 2.0

Estos valores representan los desplazamientos donde la fuerza neta sobre el resorte es cero.

## Conclusión

Los métodos cerrados, especialmente cuando se combinan con técnicas gráficas, son herramientas poderosas para encontrar raíces de ecuaciones. Sus principales ventajas son:

- Garantía de convergencia si existe una raíz en el intervalo inicial.
- Facilidad de implementación y comprensión.
- Robustez frente a funciones complejas o mal condicionadas.

Sin embargo, también tienen limitaciones:

- Pueden ser lentos comparados con métodos más avanzados.
- Requieren que la función cambie de signo en el intervalo.
- No son eficientes para encontrar raíces múltiples o muy cercanas entre sí.

En la práctica de la ingeniería, los métodos cerrados a menudo se usan como punto de partida o como complemento de métodos más avanzados, proporcionando estimaciones iniciales robustas que luego pueden refinarse con técnicas más rápidas.
