# Diferencia entre Errores Reales y Errores Percibidos

En el análisis numérico, es crucial distinguir entre **errores que existen objetivamente** y aquellos que **percibimos subjetivamente**. Esta distinción es fundamental para la interpretación precisa de resultados y la toma de decisiones informadas.

## 1. Errores Reales

### Definición
Diferencias objetivas y cuantificables entre un valor exacto (o teórico) y un valor calculado o medido.

### Características
- **Origen**: 
  * Imprecisiones en la recopilación de datos
  * Limitaciones en técnicas de medición
  * Errores de redondeo en computadoras
  * Aproximaciones en métodos numéricos

- **Impacto**: 
  * Afectan directamente la validez y precisión de los resultados
  * Pueden propagarse y amplificarse en cálculos subsecuentes

### Ejemplo
Calcular π usando la serie de Leibniz hasta el 5º término:

```python
import math

def leibniz_pi(terms):
    return sum((-1)**n / (2*n + 1) for n in range(terms)) * 4

calculated_pi = leibniz_pi(5)
real_error = abs(calculated_pi - math.pi)

print(f"Valor calculado de π: {calculated_pi}")
print(f"Valor real de π: {math.pi}")
print(f"Error real: {real_error}")
```

## 2. Errores Percibidos

### Definición
Discrepancias que los observadores creen que existen, pero que no representan una diferencia objetiva en los valores.

### Características
- **Origen**:
  * Representación visual inadecuada de datos
  * Ambigüedades en la presentación de resultados
  * Sesgos cognitivos y expectativas previas

- **Impacto**:
  * Pueden llevar a interpretaciones erróneas
  * Influyen en la toma de decisiones, incluso cuando los datos son correctos

### Ejemplo
Visualización engañosa de datos:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(x, y)
ax1.set_ylim(-0.1, 0.1)  # Escala exagerada
ax1.set_title("Percepción exagerada de variación")

ax2.plot(x, y)
ax2.set_ylim(-1, 1)  # Escala correcta
ax2.set_title("Representación adecuada")

plt.tight_layout()
plt.show()
```


## Conclusión

- Los **errores reales** requieren corrección o consideración explícita en el análisis.
- Los **errores percibidos** demandan una comunicación clara y una presentación cuidadosa de los datos.
- Comprender esta distinción es crucial para:
  1. Interpretar correctamente los resultados numéricos
  2. Tomar decisiones informadas basadas en datos
  3. Comunicar efectivamente los hallazgos científicos y técnicos
